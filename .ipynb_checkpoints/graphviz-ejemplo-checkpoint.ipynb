```python
from graphviz import Digraph, Graph
from IPython.display import display

nodos = [{'valor':'B', 'id': 1, 'parent': 0},
         {'valor': 'A', 'id': 0, 'parent': None},
         {'valor': 'C', 'id': 2, 'parent': 0}]


class Despliegue:
  def __init__(self, listado_nodos, *args, **kwargs):
    self.nodos = listado_nodos
    self.actuales = None
    self.profundidad = None
    self.hojas = None
    self.arbol = Digraph()
    
  def obtener_raiz(self):
    nodos = (filter(lambda x: x['parent'] is None , self.nodos))
    return next(nodos)

  def obtener_hijos(self, nodo):
    hijos = nodo.hijos
    self.agregar_nodos(nodo, hijos)
    
```

```python
ejemplo = Despliegue(nodos)
```

```python
ejemplo.obtener_raiz()
```

```python
p = Graph(name='parent', engine='dot')
p.edge('spam | algo', 'eggs')

with p.subgraph(name='child', node_attr={'shape': 'box'}) as c:
  c.edge('foo', 'bar')
```
